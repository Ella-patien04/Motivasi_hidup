```python
# Struktur graf menggunakan dictionary
graph = {
    "Jazz": ["Smooth Jazz", "Bebop"],
    "Smooth Jazz": ["Song1", "Song2"],
    "Bebop": ["Song3", "Song4"],
    "Rock": ["Hard Rock", "Alternative Rock"],
    "Hard Rock": ["Song5", "Song6"],
    "Alternative Rock": ["Song7", "Song8"],
    "Pop": ["Pop Rock", "Dance Pop"],
    "Pop Rock": ["Song9", "Song10"],
    "Dance Pop": ["Song11", "Song12"]
}

# Fungsi DFS untuk mencari lagu dalam subgenre tertentu
def dfs(graph, start, target):
    visited = set()  # Set untuk melacak node yang telah dikunjungi
    stack = [start]  # Menggunakan stack untuk DFS

    while stack:
        node = stack.pop()  # Ambil node terakhir dari stack
        if node not in visited:
            print(f"Visiting: {node}")  # Menampilkan proses kunjungan
            visited.add(node)

            # Jika target ditemukan, hentikan pencarian
            if node == target:
                print(f"{target} ditemukan!")
                return

            # Tambahkan tetangga dari node saat ini ke stack
            if node in graph:
                stack.extend(graph[node][::-1])  # Tambahkan secara terbalik agar DFS bekerja dengan benar

    print(f"{target} tidak ditemukan dalam graf.")

# Contoh penggunaan
# Mencari lagu di subgenre "Bebop" dalam genre "Jazz"
dfs(graph, "Jazz","Song3")
```
